```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/yanndupis/Building-a-Student-Intervention-System/master/student-data.csv')
```

```python
df
```

```python
X=df.drop(['passed'],axis=1)
```

```python
y=df['passed']
from sklearn.preprocessing import LabelEncoder
y_le=LabelEncoder().fit_transform(y)
```

```python
X_onehot=pd.get_dummies(X)
```

```python
from sklearn.preprocessing import MinMaxScaler
X_std=MinMaxScaler().fit_transform(X_onehot)
```

```python
X_std=pd.DataFrame(X_std,columns=X_onehot.columns)
X_std
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

parameter_candidates = [
  {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
  {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']},
]

clf = GridSearchCV(estimator=svm.SVC(), param_grid=parameter_candidates, n_jobs=-1)

classifier_pipeline = make_pipeline(SelectKBest(f_classif, k=10), clf)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classifier_pipeline, X_std, y_le, cv=3)
```

```python
scores
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_std,y)

classifier_pipeline.fit(X_test,y_test)
y_pred=classifier_pipeline.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

parameter_candidates = [
  {'n_estimators': [ 10, 100, 1000], 'criterion': ['gini','entropy'],'max_depth':[10,100,1000]},
]

clf = GridSearchCV(estimator=RandomForestClassifier(), param_grid=parameter_candidates, n_jobs=-1)

classifier_pipeline = make_pipeline(SelectKBest(f_classif, k=10), clf)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(classifier_pipeline, X_std, y_le, cv=3)
```
